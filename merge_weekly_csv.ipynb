# Merge Weekly Review Files

This notebook merges weekly Google Play review CSV files generated via scheduled scraping
into a single consolidated dataset. The merged dataset will be used for downstream
exploratory data analysis (EDA).

```python
import pandas as pd
from pathlib import Path

data_dir = Path("output/schedule/weekly")
csv_files = sorted(data_dir.glob("*.csv"))

len(csv_files)
```

```python
dfs = []

for f in csv_files:
    df = pd.read_csv(f)
    df["source_file"] = f.name
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.shape
```

```python
merged_df.to_csv("output/merged_chatgpt_weekly.csv", index=False)
```
